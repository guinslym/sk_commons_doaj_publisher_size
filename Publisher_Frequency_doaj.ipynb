##Frequency for 2015

```python
import pandas as pd
from pandas import DataFrame
from pandas import Series
```

```python
doaj_2015 = pd.read_csv('doaj_20151015_1800_utf8.csv', encoding='utf-8', dtype='unicode')
publishers = doaj_2015['Publisher'].value_counts()
```

```python
publishers = DataFrame(publishers.reset_index())
publishers.columns = ['Publisher','Frequency_2015']
```

```python
publishers.head(20)
```

```python
publishers.to_excel('publisher.xlsx', 
                 sheet_name = 'publisher frequencies', 
                 index = False)
```
